### 1.4.Pie plot

pie plot is basically different from the others, this only need 1 type of data, and this will created by the percentage of the value, if being a number array, it will demonstate a percentage of each elements for each types, but for a strings arrays, it will account the percentage of the amount of occurances of the elements

````python
array_number = np.abs(np.random.normal(loc=2, scale=3.5, size=random.randint(5,10))*100).astype(np.float64)
array_number
plt.pie(array_number)
````

Firstly, these are the simple pie charts parameters
- plt.pie(data, labels=, startangle =, autopct =, explode=)
  + data: given data to insert into the plot
  + labels =: given labels for each part of the graph, can be a list
  + startangle= the angle the starting part
  + autopct = a way to insert the percentage into each pie pieces of the plot, or any other elements into it, further research must will be tested
  + explode =: make each part of the graph explode

````python
##creating labels based on the piechard number of elements
Explosion = np.abs(np.random.normal(loc=0.05, scale=0.05, size=len(array_number))).astype(np.float64)
def finding_array_labels(array):
  return [f'Category of {i+1}' for i in range(len(array))] #this basicallY means that this would create a list of names with the same trends of category (number)

plt.pie(array_number,labels = finding_array_labels(array_number), startangle = -60, autopct= '%1.1f%%', explode= Explosion)
#we got the data, labels are being taken from the previous loop, startangle will be rotate to the right as it is -60, and explode are just a bunch of really small numbers
````

as for the autopct, certain way should be known and addressed, such as certain autopct value
- '%1.1f%%' means there is one digit befre and one digit after teh decimal point,
- for %% means that iff escape from the %...% then you got the actual % at the end
- nevertheless, there are certain ways to customize your shit

````python
def format_autopct(pct, all_vals): #this time, we will use lambda function on the autopct itself, so pct will be considered as given, since consideing autopct = lambda pct: format_autopct(pct,data)
  absoulute_values = int(round((pct/100)*sum(all_vals))) #which means that the input pct has been given my the library Matplotlib it self, so we onlyneed to do things with the all_vals
  return f"{pct:.1f}%\n(${absoulute_values})" #since pct that matplotlib give only shows the percentile (number only) so we have to specify it with pct/100, and then return with a % behind {}
  # and at the {pct:.1f}, the : means for the distinctions between data and data modifier, which means that .1 means it will takes one number at the decimal place

print(finding_array_labels(array_number))
plt.pie(array_number, startangle = -60, labels=finding_array_labels(array_number) , autopct= lambda pct: format_autopct(pct, array_number), explode= Explosion)
````

### 1.5. histogram


A histogram is a graphical representation of the distribution of numerical data.  It's essentially a bar chart where the x-axis represents the bins (intervals or ranges) of the data, and the y-axis represents the frequency or count of data points falling within each bin.  Histograms are used to visualize the underlying probability distribution of a dataset.  The area of each bar is proportional to the frequency of the data within that bin.

Here's how a histogram differs from a bar chart:

* **Data Representation:** Bar charts represent categorical data, with each bar corresponding to a distinct category.  Histograms represent numerical data, showing the distribution of values across continuous ranges (bins).

* **X-Axis:**  The x-axis of a bar chart represents categories (e.g., colors, countries, products).  The x-axis of a histogram represents continuous numerical ranges (bins) of the data. The bars in a histogram are adjacent, touching each other, showing the continuity of the data. There is no gap between bins.

* **Y-Axis:** The y-axis of a bar chart represents the frequency or value of each category. The y-axis of a histogram represents the frequency (or relative frequency, density, etc.) of the data points falling into each bin.

* **Purpose:** Bar charts show comparisons among discrete categories. Histograms show the shape of the distribution of numerical data.

In summary, if your data is numerical and you want to understand its distribution (how often different values occur), use a histogram. If your data is categorical and you want to compare the frequencies or values of different categories, use a bar chart.


````python
#so a histogram only receive 1 single variable, then y will means the freuency of each x range
array_number_x = np.abs(np.random.normal(loc=5000, scale=1000, size=1000))
plt.hist(array_number)
````

### 2.2 Multiple plots in one figure by twinx()

- using twinx() means that it will creates another axis or just an extra sides on a whole figure of oxy, for instance, theoretically, when we finished executing twinx(), we got x line, y line like a traditional figure, but we also got a self-established y1 on the verticle right with the y on the verticle left in tandem. which means this particular graph will represents 2 differents sets of xs and ys
- to execute twinx(), we have to execute


```
fig, ax_1 = plt.subplots()

```
- after that, we can established our traditional graph

```
ax_1.plot(months, temp_c)
ax_1.set_ylabel('Temperature of Celsius')
```
- and then to add another verticle axis, we add


```
ax_2 = ax_1.twinx()
ax_2.set_ylabel('Temperature of Fahrenheit')
ax_2.set_ylim(60, 90);
```





````python
temp_c = np.array([18,20,21,23,28,30,30,28,28,25,21,16])
temp_f = temp_c*1.8 + 32
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']


fig, ax_1 = plt.subplots()
ax_1.plot(months, temp_c)
ax_1.set_ylabel('Temperature of Celsius')

ax_2 = ax_1.twinx()
ax_2.set_ylabel('Temperature of Fahrenheit')
ax_2.set_ylim(60, 90);
````
