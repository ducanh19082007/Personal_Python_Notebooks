##2. Pandas Series

- what is a series anyway?
- a pandas series is like a column in a table
- and a one-dimensional array holding data of any type
- and it is like a way to change a dictionary into a dataframe-like way, and it can in fact change into a DataFrame after all
- Nevertheless, a series was create but in a single column of 1D, while a DataFrame is a table of 2-D
- Thus, sometime, using series is actually really simple and convinient, since it is simpler, faster
- Deeper analysis, we can seperate them into the need of using case
 + pandas.series can use to storing and manipulating ***single columns or lists***
 + while pandas.DataFrame can be use to handling multiple columns and rows of data

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.Series(data)

print(myvar) #a dataframe in a form of a series
```

```python
pd.DataFrame(myvar)
```

```python
pd.DataFrame(data)
```

```python
import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```

```python

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ['day1','day3'])

print(myvar)
```
